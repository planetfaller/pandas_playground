```python
df = pd.read_csv('daily-minimum-temperatures-in-me.csv')
df.head()
df.Date = pd.to_datetime(df.Date)
df.set_index('Date',inplace=True)
df.head()
```

```python
df.rolling(365).mean().plot(figsize=(20,10))
```

```python
df.plot(figsize=(20,10), style='.' )
```

```python
df.diff().plot(figsize=(20,10)) #difference plot first order
```

```python
pd.plotting.lag_plot(df) #data with autocorrelation gives rise to linear patters that follow the diagonal
```

```python
pd.plotting.autocorrelation_plot(df)
```

```python
x = np.linspace(0,100*math.pi,num=1000)
y = np.sin(x)
```

```python
df_sin = pd.DataFrame({'x': x,'y':y})
df_sin.head()
```

```python
df_sin[0:1000].plot(x='x',y='y',figsize=(15,3))
```

```python
pd.plotting.lag_plot(df_sin.y)
```

```python
pd.plotting.autocorrelation_plot(df_sin.y)
```
