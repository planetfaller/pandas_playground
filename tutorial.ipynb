```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
sns.set()#?
```

```python
df = pd.read_csv('data/multiTimeline.csv', skiprows=1)
df.head()
```

```python
df.columns = ['month','diet','gym','finance']
df.head()
```

```python
df.month = pd.to_datetime(df.month)
df.set_index('month',inplace=True)
df.head()
```

```python
df.plot(figsize=(20,10), linewidth=5,fontsize=20) #plot all the vars
plt.xlabel('Year',fontsize=20)
```

```python
#plotting diet by itself
diet = df[['diet']] #two brackets to extract column, one to create pandas series
diet.plot(figsize = (20,10))
```

```python
diet.rolling(12).mean().plot(figsize=(20,10)) #rolling avg with window, we remove some of the seasonality and see trends over years
```

```python
gym = df[['gym']]
gym.rolling(12).mean().plot(figsize=(20,10))
```

```python
df_rm = pd.concat([diet.rolling(12).mean(), gym.rolling(12).mean()], axis=1) #concatenate use axis 1
df_rm.plot(figsize=(20,10))
```

```python
diet.diff().plot(figsize=(20,10)) # look at first order difference, jan sales trend clearly visible
```

```python
from sklearn import datasets
iris = datasets.load_iris()
df_iris = pd.DataFrame(data= np.c_[iris['data'], iris['target']], columns= iris['feature_names'] + ['target']) #data formating 
df_iris.head()
```

```python
sns.lmplot(x='sepal length (cm)', y='sepal width (cm)',data=df_iris, fit_reg=False,hue='target') #scatter plot of length vs width, regression can be turned on, hue based on target
```

```python
df_iris.corr() #lets run correlation
```
